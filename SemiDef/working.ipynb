```python
import numpy as np

matrix = np.zeros((6, 6))
np.fill_diagonal(matrix, -1.0)
np.fill_diagonal(np.fliplr(matrix), 1000.0)

print(matrix)
print(matrix)

eigenvalues = np.linalg.eigvalsh(matrix)

print("Eigenvalues (sorted ascending):")
for i, val in enumerate(eigenvalues):
    print(f"  λ{i+1} = {val: .8f}")

print(f"\nPositive semi-definite: {np.all(eigenvalues >= -1e-10)}")
print(f"Min eigenvalue: {eigenvalues.min():.8f}")
print(f"Max eigenvalue: {eigenvalues.max():.8f}")
print(f"Trace (sum of eigenvalues): {eigenvalues.sum():.8f}")
```
