```python
# Import packages.
import cvxpy as cp
import numpy as np
```

```python
A1 = [[1, 0, 1], [0, 3, 7], [1, 7, 5] ]
A1 = np.array(A1)

A2 = [[0,2,8], [2, 6, 0], [8,0,4]]
A2 = np.array(A2)

C = [[1,2,3,], [2,9,0],[3,0,7]]
C = np.array(C)

b1 = 11
b2 = 19

X = cp.Variable((3,3), symmetric=True)
```

```python
constraints = [X >> 0] 
constraints += [(cp.vdot(A1, X) == b1)]
constraints += [(cp.vdot(A2, X) == b2)]
```

```python
prob = cp.Problem(cp.Minimize(cp.vdot(C, X)), constraints)
prob.solve()

# Print result.
print("The optimal value is", prob.value)
print("A solution X is")
print(X.value)
```
